```python
# trip updates
url = f'http://api.511.org/transit/vehiclepositions?api_key={api_key}&agency={operator}'
response = requests.get(url)
response.raise_for_status() 
```

```python
# Parse the Protocol Buffer data
feed = gtfs_realtime_pb2.FeedMessage()
feed.ParseFromString(response.content)

# Convert to dictionary
data = MessageToDict(feed)
```

```python
vehicle_positions = []
for entity in data['entity']:
    if 'vehicle' in entity:
        vehicle = entity['vehicle']
        vehicle_positions.append({
            'trip_id': vehicle['trip']['tripId'],
            'route_id': vehicle['trip']['routeId'],
            'vehicle_id': vehicle['vehicle']['id'],
            'current_stop_sequence': vehicle.get('currentStopSequence', None),
            'current_status': vehicle.get('currentStatus', None),
            'timestamp': vehicle.get('timestamp', None),
            'latitude': vehicle['position']['latitude'],
            'longitude': vehicle['position']['longitude'],
            'bearing': vehicle['position'].get('bearing', None),
            'speed': vehicle['position'].get('speed', None),
        })

# Convert to DataFrame for easier manipulation
vehicle_positions_df = pd.DataFrame(vehicle_positions)

print(vehicle_positions_df.head())
```
